```python
import json
from getpass import getpass
import time

f = open('bank_problem.db','r')
bank=json.load(f)
f.close()

Admin = 'babe'
Admin_password = 'wonderlAnd'
while True:
    print("\n\n")
    print("*"*60)
    print("\t\t",time.ctime())
    print("\t\tWELCOME TO PYTHON BANK")
    print("*"*60)
    x = int(input("Enter your choice : \n1. Login\n2. Signup\n3. Admin\n4. Exit\n\n"))
    if x == 1:
        uacc = int(input("\n\nEnter Your Account Number : "))
        if uacc in bank['acc']:
            upass = getpass("\nEnter the password : ")
            index = bank['acc'].index(uacc)
            
            f = open('log.txt','a')
            s = "\n User with Account no. {} has logged in at {}".format(uacc,time.ctime())
            f.write(s)
            f.close()
            
            if upass == bank['password'][index]:
                while True:
                    print("\n\n1. Debit\n2. Credit\n3. Check Balance\n4. Change Password\n5. Log out\n")
                    a = int(input("\nEnter your Choice : "))
                    if a == 1:
                        d = int(input("\n\nHow much money you want to debit : "))
                        if d > bank['balance'][index]:
                            print("\n\nInsufficient Balance to withdraw!")
                            print("\nYou have {} amount in your account".format(bank['balance'][index]))
                        else:
                            bank['balance'][index] = bank['balance'][index] - d
                            print("\n\nYou have left {} rupees in your account now after debitting".format(bank['balance'][index]))
                            
                            f = open('log.txt','a')
                            s = "\nUser with account number {} had withdrew {} money at {}".format(uacc,bank['balance'][index],time.ctime())
                            f.write(s)
                            f.close()
                            
                            f = open('bank_problem.db','w')
                            json.dump(bank,f)
                            f.close()
                    
                    elif a == 2:
                        c = float(input("\n\nHow much money you want to credit?"))
                        bank['balance'][index] = bank['balance'][index] + c
                        print("\nYou have {} balance in your account after creditting".format(bank['balance'][index]))
                        
                        f = open('log.txt','a')
                        s = "\nUser with Account number {} had creditted {} money at {}".format(uacc,bank['balance'][index],time.ctime())
                        f.write(s)
                        f.close()
                        
                        f = open('bank_problem.db','w')
                        json.dump(bank,f)
                        f.close()
                    
                    elif a == 3:
                        print("\nYou have {} balance in your account".format(bank['balance'][index]))
                        
                        f = open('log.txt','a')
                        s = "\nUser with Account number {} had checked his balance which is {} at {}".format(uacc,bank['balance'][index],time.ctime())
                        f.write(s)
                        f.close()
                    
                    elif a == 4:
                        op = input("\n\nEnter you old password : ").strip()
                        if op == bank['password'][index]:
                            np = getpass("\nEnter a new password = ").strip()
                            cp = getpass("Reenter your password = ").strip()
                            if np == cp:
                                bank['password'][index] = np
                                print("\n\nPASSWORD SUCCESSFULLY CHANGED")
                                
                                f = open('log.txt','a')
                                s = "\nUser with Account number {} had changed his old password to {} at {}".format(uacc,np,time.ctime())
                                f.write(s)
                                f.close()
                                
                                f = open('bank_problem','w')
                                json.dump(bank,f)
                                f.close()
                            
                            else:
                                print("\nPASSWORD DOES NOT MATCH! TRY AGAIN!")
                                
                                f = open('log.txt','a')
                                s = "\nUser with Account number {} had entered {} while rentering the password {} at {}".format(uacc,cp,np,time.ctime())
                                f.write(s)
                                f.close()
                        
                        else:
                            print("\n\nINCORRECT OLD PASSWORD!")
                            
                            f = open('log.txt','a')
                            s = "\nUser with Account number {} had entered worng old password which was {} at {}".format(uacc,bank['password'][index],time.ctime())
                            f.write(s)
                            f.close()
                    
                    elif a == 5:
                        print("\n\nLOGGING YOU OUT.....")
                        break
                        
                        f = open('log.txt','a')
                        s = "\nUser with Account number {} had logged out at {}".format(uacc,time.ctime())
                        f.write(s)
                        f.close()
                    
                    else:
                        print("\nINVALID CHOICE")
                        
                        f = open('log.txt','a')
                        s = "\nUser with Account number {} had entered wrong choice at {}".format(uacc,time.ctime())
                        f.write(s)
                        f.close()
            
            else:
                print("\nINCORRECT PASSWORD!")
                
                f = open('log.txt','a')
                s = "\nUser with Account number {} had entered wromg password at {}".format(uacc,time.ctime())
                f.write(s)
                f.close()
        
        else:
            print("\n\nINCORRECT ACCOUNT NUMBER!")
            
            f = open('log.txt','a')
            s = "\nUser had entered incorrect Account number at {}".format(time.ctime())
            f.write(s)
            f.close()
    
    elif x == 2:
        acc = bank['acc'][-1]+1
        name= input("\nEnter your username = ").strip().upper()
        bal = float(input("Enter your Balance = "))
        password = getpass("Enter password")
        bank['user'].append(name)
        bank['balance'].append(bal)
        bank['password'].append(password)
        bank['acc'].append(acc)
        print("\n\nACCOUNT IS SUCCESSFULLY MADE!")
        print("\nYour Account number is : {}".format(acc))
        
        f = open('log.txt','a')
        s = "\nA new User had opened an Account with account number {}, username {}, with balance {}, and password {} at {}".format(acc,name,bal,password,time.ctime())
        f.write(s)
        f.close()
        
        f = open('bank_problem.db','w')
        json.dump(bank,f)
        f.close()
    
    elif x == 3:
        ad = input("\nEnter Admin Name : ").strip().lower()
        if ad == 'babe':
            pd = getpass("\nEnter Admin Password : ").strip()
            if pd == 'wonderlAnd':
                for key,value in bank.items():
                    print("{} = {}".format(key,value))
                    print()
                    
                    f = open('log.txt','a')
                    s = "\nAdmin Had logged in at {}".format(time.ctime())
                    f.write(s)
                    f.close()
            
            else:
                print("\n\nINCORRECT PASSWORD")
                
                f = open('log.txt','a')
                s = "\nAmin had entered incorrect password at {}".format(time.ctime())
                f.write(s)
                f.close()
        
        else:
            print("\n\nINCORRECT ADMIN NAME")
            
            f = open('log.txt','a')
            s = "\nAdmin had entered incorrect Username at {}".format(time.ctime())
            f.write(s)
            f.close()
    
    elif x == 4 :
        e = input("\n\nAre You Sure You Want To Exit?\n(y/n) : ").strip().lower()
        if e == 'y' or e == 'yes' :      
            print("\nTHANK YOU FOR USING OUR SERVICES ")
            break
            
            f = open('log.txt','a')
            s = "\nUser Chose to Exit at {}".format(time.ctime())
            f.write(s)
            f.close()
    
    else:
        print("\n\nENTER CORRECT CHOICE")
        
        f = open('log.txt','a')
        s = "\nUser entered wrong choice at {}".format(time.ctime())
        f.write(s)
        f.close()
        
```
